```python
#vars
gama = 1.4
nx = 81
dx = .25
dt = .0002
nt = int(.01/dt)+1
L=10.

x = numpy.linspace(-L,L,nx)
u = numpy.zeros((1,nx,3))
#u.shape
```

```python
####Initial Condition for both LHS and RHS####
L_density_initial = 1.
L_velocity_initial = 0.
L_pressure_initial = 100000.
R_density_initial = 0.125
R_velocity_initial = 0.
R_pressure_initial = 10000.
    
Lu1 = L_density_initial 
Lu2 = L_density_initial*L_velocity_initial
Lu3 = L_density_initial*(L_pressure_initial/((gama-1.)*L_density_initial)+.5*(L_velocity_initial)**2) 
    
Ru1 = R_density_initial 
Ru2 = R_density_initial*R_velocity_initial
Ru3 = R_density_initial*(R_pressure_initial/((gama-1.)*R_density_initial)+.5*(R_velocity_initial)**2)
 
lic = [Lu1,Lu2,Lu3];    
ric =[Ru1,Ru2,Ru3];
  
u[0,0:40,:] = lic;
u[0,40:81,:] = ric;
```

```python
def computeF(u,gama):
    F = numpy.zeros((1,nx,3))
   # print(F.shape)
   # print(u.shape)
    F[0,:,0] = u[0,:,1]
    F[0,:,1] = (u[0,:,1]**2)/u[0,:,0] + (gama-1.)*(u[0,:,2]-(.5*(u[0,:,1]**2)/u[0,:,0]))
    F[0,:,2] = (u[0,:,2]+(gama-1.)*(u[0,:,2]-.5*(u[0,:,1]**2/u[0,:,0])))*(u[0,:,1]/u[0,:,0])
    
    return F

#F = numpy.zeros((nx,3))
#for i in range(nx):
#    F[i,:] = computeF(u[i,:],gama)
#computeF(u,gama,nx)
```

```python
def computeFstar(ustar,gama):
    Fstar = numpy.zeros((1,nx,3))
    Fstar[0,:,0] = ustar[0,:,1]
    Fstar[0,:,1] = ustar[0,:,1]**2/ustar[0,:,0]+(gama-1.)*(ustar[0,:,2]-.5*(ustar[0,:,1]**2/ustar[0,:,0]))
    Fstar[0,:,2] = (ustar[0,:,2]+(gama-1.)*(ustar[0,:,2]-.5*(ustar[0,:,1]**2/ustar[0,:,0])))*(ustar[0,:,1]/ustar[0,:,0])
    
    return Fstar
```

```python
def richtmyer(u, nt, dt, dx, nx): 
    damp = 0.1
    u_n = numpy.zeros((nt,nx,3))
    u_n[:,:,:] = u.copy()
    u_star = u.copy()
    
    for t in range(1,nt):
        #print(u.shape)
        F = computeF(u,gama)
        
        u_star[0,2:,:] = .5*(u[0,2:,:] + u[0,1:-1,:]) - dt/(2 * dx) * (F[0,2:,:]-F[0,1:-1,:])
        
        Fstar = computeF(u_star,gama)
        
        u_n[t,1:-1,:] = u[0,1:-1,:] - (dt/dx) * (Fstar[0,2:,:] - Fstar[0,1:-1,:]) +(damp)*(u[0,2:,:] -2*u[0,1:-1,:] +u[0,:-2,:])

        u[:,0,:] = u[:,1,:]  #since we don't have reflective BC, this is the anti-blowup code
        u[:,-1,:] = u[:,-2,:] #this one, too
        
        u[0,:,:] = u_n[t,:,:].copy()

    return u_n
```

```python
u_n = richtmyer(u,nt,dt,dx,nx)
```
