```python
import math;
import numpy;
import sympy;
import matplotlib.pyplot as plt;
%matplotlib inline
from matplotlib import animation
from JSAnimation.IPython_display import display_animation
```

```python
#vars
gama = 1.4
nx = 81
dx = .25
dt = .0002
nt = int(.01/dt)+1
L=10.

x = numpy.linspace(-L,L,nx)
u = numpy.zeros((1,nx,3))
```

```python
####Initial Condition for both LHS and RHS####
L_density_initial = 1.
L_velocity_initial = 0.
L_pressure_initial = 100000.
R_density_initial = 0.125
R_velocity_initial = 0.
R_pressure_initial = 10000.
    
Lu1 = L_density_initial 
Lu2 = L_density_initial*L_velocity_initial
Lu3 = L_density_initial*(L_pressure_initial/((gama-1.)*L_density_initial)+.5*(L_velocity_initial)**2) 
    
Ru1 = R_density_initial 
Ru2 = R_density_initial*R_velocity_initial
Ru3 = R_density_initial*(R_pressure_initial/((gama-1.)*R_density_initial)+.5*(R_velocity_initial)**2)
 
lic = [Lu1,Lu2,Lu3];    
ric =[Ru1,Ru2,Ru3];
  
u[0,0:40,:] = lic;
u[0,40:81,:] = ric;
```

```python
def computeF(u,gama):
    F = numpy.zeros((1,nx,3))
   # print(F.shape)
   # print(u.shape)
    F[0,:,0] = u[0,:,1]
    F[0,:,1] = (u[0,:,1]**2)/u[0,:,0] + (gama-1.)*(u[0,:,2]-(.5*(u[0,:,1]**2)/u[0,:,0]))
    F[0,:,2] = (u[0,:,2]+(gama-1.)*(u[0,:,2]-.5*(u[0,:,1]**2/u[0,:,0])))*(u[0,:,1]/u[0,:,0])
    
    return F
```

```python
def richtmyer(u, nt, dt, dx, nx): #, damp = 0.05
    
    u_n = numpy.zeros((nt,nx,3))
    u_n[:,:,:] = u.copy()
    u_star = u.copy()
    

    for t in range(1,nt):
        #print(u.shape)
        F = computeF(u,gama)
        
        u_star[0,1:,:] = .5*(u[0,1:,:] + u[0,:-1,:]) - dt/(2 * dx) * (F[0,1:,:]-F[0,:-1,:])
        
        Fstar = computeF(u_star,gama)
        
        u_n[t,0:-1,:] = u[0,:-1,:] - (dt/dx) * (Fstar[0,1:,:] - Fstar[0,:-1,:]) #+ damp * ()
        
        u[:,0,:] = u[:,1,:]  #since we don't have reflective BC, this is the anti-blowup code
        u[:,-1,:] = u[:,-2,:] #this one, too
        
        u[0,:,:] = u_n[t,:,:].copy()

    return u_n
```

```python
u_n = richtmyer(u,nt,dt,dx,nx)
```

```python
fig = plt.figure();
ax = plt.axes(xlim=(-L,L),ylim=(0,1.5),xlabel=('Distance'),ylabel=('Traffic density'));
line, = ax.plot([],[],color='#003366', lw=2);

def animate(frame):
    x = numpy.linspace(-L,L,nx)
    y = u_n[frame,:,0]
    line.set_data(x,y)
    return line,

anim = animation.FuncAnimation(fig, animate, frames=nt-1, interval=50)
display_animation(anim, default_mode='once')
```
