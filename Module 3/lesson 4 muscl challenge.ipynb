```python
%matplotlib inline
import numpy 
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
from matplotlib import animation
from JSAnimation.IPython_display import display_animation
```

```python
#vars
gama = 1.4
nx = 81
dx = .25
dt = .0002
nt = int(.01/dt+1)
L=10.

x = numpy.linspace(-L,L,nx)
u = numpy.zeros((nx,3))  #matrix for IC's
```

```python
####Initial Condition for both LHS and RHS####
L_density_initial = 1.
L_velocity_initial = 0.
L_pressure_initial = 100000.
R_density_initial = 0.125
R_velocity_initial = 0.
R_pressure_initial = 10000.
    
Lu1 = L_density_initial 
Lu2 = L_density_initial*L_velocity_initial
Lu3 = L_density_initial*(L_pressure_initial/((gama-1.)*L_density_initial)+.5*(L_velocity_initial)**2) 
    
Ru1 = R_density_initial 
Ru2 = R_density_initial*R_velocity_initial
Ru3 = R_density_initial*(R_pressure_initial/((gama-1.)*R_density_initial)+.5*(R_velocity_initial)**2)
 
lic = [Lu1,Lu2,Lu3];    
ric =[Ru1,Ru2,Ru3];
  
u[0:40,:] = lic;
u[40:81,:] = ric;
```

```python
#IC Plots

subplot(3,3,1)
plt.plot(x,u[:,0])
plt.xlabel('position (m)', fontsize=14)
plt.ylabel('u_1', fontsize=14)
plt.grid(True)
plt.title('Initial Condition u_1', fontsize=16);

subplot(3,3,3);
plt.plot(x,u[:,1])
plt.xlabel('position (m)', fontsize=14)
plt.ylabel('u_2', fontsize=14)
plt.grid(True)
plt.title('Initial Condition u_2', fontsize=16);

subplot(3,3,8);

plt.plot(x,u[:,2])
plt.xlabel('position (m)', fontsize=14)
plt.ylabel('u_3', fontsize=14)
plt.grid(True)
plt.title('Initial Condition u_3', fontsize=16);
```

```python
def computeF(u,gama):
    F = numpy.zeros((nx,3))
    F[:,0] = u[:,1]
    F[:,1] = (u[:,1]**2)/u[:,0] + (gama-1.)*(u[:,2]-(.5*(u[:,1]**2)/u[:,0]))
    F[:,2] = (u[:,2]+(gama-1.)*(u[:,2]-.5*(u[:,1]**2/u[:,0])))*(u[:,1]/u[:,0])
    
    return F
```

```python
def minmod(e,nx,dx):
    """
    Compute the minmod approximation to the slope
    
    Parameters
    ----------
    e  : array of float 
        input data
    dx : float 
        spacestep
    nx : int
        number of spacial steps
    
    Returns
    -------
    sigma : array of float 
            minmod slope
    """
    
    sigma = numpy.zeros((nx,3))
    de_minus = numpy.zeros((nx,3))
    de_plus = numpy.zeros((nx,3))
    
    de_plus.shape
    
    de_minus[1:,:] = (e[1:,:] - e[:-1,:])/dx
    de_plus[:-1,:] = (e[1:,:] - e[:-1,:])/dx
    # The following is inefficient but easy to read
    for j in range(3):
        for i in range(1, nx-1):
            if (de_minus[i,j] * de_plus[i,j] < 0.0):
                sigma[i,j] = 0.0
            elif (numpy.abs(de_minus[i,j]) < numpy.abs(de_plus[i,j])):
                sigma[i,j] = de_minus[i,j]
            else:
                sigma[i,j] = de_plus[i,j]
    return sigma
```

```python
def muscl(u, nt, dt, dx, nx):
    """ Computes the solution with the MUSCL scheme using the Lax-Friedrichs flux,
    RK2 in time and minmod slope limiting.
    
    Parameters
    ----------
    rho    : array of floats
            Density at current time-step
    nt     : int
            Number of time steps
    dt     : float
            Time-step size
    dx     : float
            Mesh spacing
    rho_max: float
            Maximum allowed car density
    V_max  : float
            Speed limit
    
    Returns
    -------
    rho_n : array of floats
            Density after nt time steps at every point x
    """
    
    #initialize our results array with dimensions nx by nt with a vector of 3 components in each element
    u_n = numpy.zeros((nx,nt,3))      
    #copy the initial u array into each row of our new array
    for i in range(nx):
        for j in range(3):
            u_n[i,0,j]=u[i,j]
                  
    
    #setup some temporary arrays
    u_plus = numpy.zeros_like(u)
    u_minus = numpy.zeros_like(u)
    flux = numpy.zeros_like(u)
    u_star = numpy.zeros_like(u)

    for t in range(1,nt):
        sigma = minmod(u,nx,dx) #calculate minmod slope
        
        #reconstruct values at cell boundaries
        u_left = u + sigma*dx/2.
        u_right = u - sigma*dx/2.
        
        flux_left = computeF(u_left,gama)
        flux_right = computeF(u_right,gama)
        
        #flux i = i + 1/2
        flux[:-1,:] = 0.5 * (flux_right[1:,:] + flux_left[:-1,:] - dx/dt *(flux_right[1:,:] - flux_left[:-1,:]))
       
        #rk2 step 1
        u_star[1:-1,:] = u[1:-1,:] + dt/dx*(flux[:-2,:] - flux[1:-1,:])
        u_star[0,:] = u[0,:]
        u_star[-1,:] = u[-1,:]
             
        sigma = minmod(u_star,nx,dx) #calculate minmod slope
    
        #reconstruct values at cell boundaries
        u_left = u_star + sigma*dx/2.
        u_right = u_star - sigma*dx/2.
        
        flux_left = computeF(u_left,gama)
        flux_right = computeF(u_right,gama)
        
        flux[:-1,:] = 0.5 * (flux_right[1:,:] + flux_left[:-1,:] - dx/dt * (flux_right[1:,:] - flux_left[:-1,:] ))
        
        u_n[1:-1,t,:] = 0.5 * (u[1:-1,:] + u_star[1:-1,:] + dt/dx*(flux[:-2,:]-flux[1:-1,:]))
        u_n[0,t,:] = u[0,:].copy()
        u_n[-1,t,:] = u[-1,:].copy()
        u = u_n[:,t,:].copy()
        
    return u_n
```

```python
cfl = .01

dt = cfl*dx

u_n = muscl(u, nt, dt, dx, nx)
```

```python
fig = plt.figure();
ax = plt.axes(xlim=(-L,L),ylim=(0,1.5),xlabel=('Distance'),ylabel=('Traffic density'));
line, = ax.plot([],[],color='#003366', lw=2);

def animate(frame):
    x = numpy.linspace(-L,L,nx)
    y = u_n[:,frame,0]
    line.set_data(x,y)
    return line,

anim = animation.FuncAnimation(fig, animate, frames=nt, interval=50)
display_animation(anim, default_mode='once')
```
