```python
from matplotlib import animation
from JSAnimation.IPython_display import display_animation
```

```python
nx = 81
nt = 70
dx = 4.0/(nx-1)
```

```python
def u_initial():
    
    u = numpy.zeros(nx)
    u[0:2.0/dx] = 1.0
    return u
```

```python
u = u_initial()
print(u)
```

```python
computeF = lambda u: (u)**2/2
```
