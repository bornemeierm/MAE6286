```python
#inits
import numpy
import sympy
import math
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

```python
#variables
c=1.             #wavespeed

#time settings
dt = .02         #size of delta t
nt = 25         #number of steps in time. Total time stepped= .02*25 = .05

#distance settings
nx = 101      #number of spatial gridpoints
dx= 2./(nx-1)    #distance between any adjacent gridpoints
```

```python
#IC
u=numpy.ones(nx)
u[.5/dx : 1/dx+1] = 2
print(u)
```

```python
plt.plot(numpy.linspace(0,2,nx), u,'g', lw=3)
plt.ylim(0,2.5);
```

```python
for n in range(1,nt):  
    un = u.copy() 
    plt.plot(numpy.linspace(0,2,nx), u, color='#003366', ls='--',\
    lw=3, alpha=(1./nt)*n);
    plt.ylim(0,2.5);
    for i in range(1,nx): 
        u[i] = un[i]-c*dt/dx*(un[i]-un[i-1])
        
        
#plt.plot(numpy.linspace(0,2,nx), u, 'g', ls='--', lw=3)
#plt.ylim(0,2.5);
```
