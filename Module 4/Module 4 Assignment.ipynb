```python
import numpy
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
```

```python
n = 192
Du, Dv, F, k = 0.00016, 0.00008, 0.035, 0.065 # Bacteria 1 
dh = 5./(n-1)
T = 8000
dt = .9 * dh**2 / (4*max(Du,Dv))
nt = int(T/dt)
```

```python
us = numpy.empty((n,n,nt+1));
vs = numpy.empty((n,n,nt+1));

uvinitial = numpy.load('C:/Users/Matt/Documents/IPython Notebooks/MAE6286 Numerical Methods/Module 4/uvinitial.npz')
us[:,:,0] = uvinitial['U']
vs[:,:,0] = uvinitial['V']
```

```python
fig = plt.figure(figsize=(8,5))
plt.subplot(121)
plt.imshow(us[:,:,0], cmap = cm.RdBu)
plt.xticks([]), plt.yticks([]);
plt.subplot(122)
plt.imshow(vs[:,:,0], cmap = cm.RdBu)
plt.xticks([]), plt.yticks([]);
```

```python
for t in range(0,nt):
  
    us[1:-1,1:-1,t+1] = dt * (Du/(dh**2) * (us[2:,1:-1,t] - 2 * us[1:-1,1:-1,t] + us[:-2,1:-1,t] + us[1:-1,2:,t] - 2 * us[1:-1,1:-1,t] +\
        us[1:-1,:-2,t]) - us[1:-1,1:-1,t] * (vs[1:-1,1:-1,t])**2 + F * (1. - us[1:-1,1:-1,t])) + us[1:-1,1:-1,t]
    
    vs[1:-1,1:-1,t+1] = dt * (Dv/(dh**2) * (vs[2:,1:-1,t] - 2 * vs[1:-1,1:-1,t] + vs[:-2,1:-1,t] + vs[1:-1,2:,t] - 2 * vs[1:-1,1:-1,t] +\
        vs[1:-1,:-2,t]) + us[1:-1,1:-1,t] * (vs[1:-1,1:-1,t])**2 - (F + k) * vs[1:-1,1:-1,t]) + vs[1:-1,1:-1,t]
    

    us[0,1:-1,t+1]  = us[1,1:-1,t+1]
    us[-1,1:-1,t+1] = us[-2,1:-1,t+1]
    us[1:-1,0,t+1]  = us[1:-1,1,t+1]
    us[1:-1,-1,t+1]  = us[1:-1,-2,t+1]
    
    vs[0,1:-1,t+1]  = vs[1,1:-1,t+1]
    vs[-1,1:-1,t+1] = vs[-2,1:-1,t+1]
    vs[1:-1,0,t+1]  = vs[1:-1,1,t+1]
    vs[1:-1,-1,t+1]  = vs[1:-1,-2,t+1]
    
```

```python
fig = plt.figure(figsize=(8,5))
plt.subplot(121)
plt.imshow(us[:,:,-1],cmap=cm.RdBu)
plt.xticks([]), plt.yticks([]);
plt.subplot(122)
plt.imshow(vs[:,:,-1],cmap=cm.RdBu)
plt.xticks([]), plt.yticks([]);

print(us[100,::40,nt])
print(numpy.round(us[100,::40,nt],4))
```
