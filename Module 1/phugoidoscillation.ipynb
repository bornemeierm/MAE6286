```python
# parameters
g=9.81;
v_t = 30.0;
C_D = 1/40.;
C_L = 1.0

#I.C.'s
v0 = v_t
theta0 = 0.0;
x0 = 0.0;
y0 = 1000.0;
```

```python
def f(u):
    v = u[0]
    theta = u[1]
    x = u[2]
    y = u[3]
    return numpy.array([-g*sin(theta) - C_D/C_L*g/v_t**2*v**2, -g*cos(theta)/v + g/v_t**2*v, v*cos(theta), v*sin(theta)])
```

```python
def euler_step(u, f, dt):
    return u + dt * f(u)
```

```python
T = 100.
dt = 0.1
N = int(T/dt) + 1
t = numpy.linspace(0.0, T, N)
u = numpy.empty((N, 4))
u[0] = numpy.array([v0, theta0, x0, y0])

for n in range(N-1):
    
    u[n+1] = euler_step(u[n], f, dt)
```

```python
x = u[:,2]
y = u[:,3]
plt.plot(x,y)
```
